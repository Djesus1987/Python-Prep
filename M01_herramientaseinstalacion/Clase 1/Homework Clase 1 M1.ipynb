2) Convertir de decimal a binario las fracciones 1/2, 1/3, 1/4, 1/5, 1/6, 1/7, 1/8, 1/9. Luego analizar los resultados y observar qué particularidad se encuentra en los mismos. Se puede usar Python o una calculadora, lo importante es ver si hay algo que podemos notar...

```python
def decimal_a_binario(decimal):
    parte_entera = int(decimal)
    parte_decimal = decimal - parte_entera
    binario_entero = bin(parte_entera)[2:]

    resultado_decimal = ''
    while parte_decimal != 0:
        parte_decimal *= 2
        if parte_decimal >= 1:
            resultado_decimal += '1'
            parte_decimal -= 1
        else:
            resultado_decimal += '0'

    return binario_entero + '.' + resultado_decimal
```

```python
print(f'fraccion 1/2 = 0.5 en binario es={decimal_a_binario(0.5)}')
print(f'la fraccion de 1/2 es = {1/2} en binario es= {decimal_a_binario(0.3333333333333333)}')
print(decimal_a_binario(0.25))
print(decimal_a_binario(0.2))
print(decimal_a_binario(0.16666666666666666))
print(decimal_a_binario(0.14285714285714285))
print(decimal_a_binario(0.125))
print(decimal_a_binario(0.1111111111111111))

```
